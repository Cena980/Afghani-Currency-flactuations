```python
# visualize the correlation between OOR, CR, ATR, LR, and GDP, 
# create a list of relevant data
relevant_columns = ['GDP', 'CR']

# subset the dataframe
df_subset = af_df[relevant_columns]

# calculate correlation matrix
correlation_matrix = df_subset.corr(numeric_only=True)

# create correlation heatmap
plt.figure(figsize=(25, 18))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True, fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of GDP and Currency Rate For Somalia')
plt.savefig('Correlation Heatmap of GDP and Currency Rate For Somalia.png')
plt.show()
```

```python
# visualize the correlation between OOR, CR, ATR, LR, and GDP, 
# create a list of relevant data
relevant_columns = ['GDP', 'Inf', 'CR']

# subset the dataframe
df_subset = af_df[relevant_columns]

# calculate correlation matrix
correlation_matrix = df_subset.corr(numeric_only=True)

# create correlation heatmap
plt.figure(figsize=(25, 18))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True, fmt=".2f", linewidths=.5)
# plt.title("Correlation Heatmap of GDP, Inflation, and Currency Rate for Afghanistan")
plt.savefig('Correlation Heatmap of GDP, Inflation, and Currency Rate for Afghanistan.png')
plt.show()
```

As can be seen in the correlation graph, Currency rate has an indirect relationship with GDP and Inflation. The increase in the two causes decrease Currency Rate.
