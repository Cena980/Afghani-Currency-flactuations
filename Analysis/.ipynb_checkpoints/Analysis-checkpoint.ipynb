```python
#Importing required packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
af_df = pd.read_csv(r'C:\Users\Sina\Desktop\Afghani-Currency-flactuations\Datasets\Cleaning Process\Afghanistan2.csv')
```

```python
# visualize the correlation between OOR, CR, ATR, LR, and GDP, 
# create a list of relevant data
relevant_columns = ['GDP', 'Inf', 'CR']

# subset the dataframe
df_subset = af_df[relevant_columns]

# calculate correlation matrix
correlation_matrix = df_subset.corr(numeric_only=True)

# create correlation heatmap
plt.figure(figsize=(25, 18))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True, fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of GDP, Inflation, and Currency Rate')
plt.savefig('Correlation Heatmap of GDP, Inflation, and Currency Rate.png')
plt.show()
```
